```python
data = pd.get_dummies(data, columns=["Fuel"], drop_first=True)
```

```python
X = data.drop('Price', axis=1)
Y = data['Price']
X_train, X_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
X_train, X_test, y_train, y_test
```

```python
models = {}
```

```python
lr1 = LinearRegression()
lr1.fit(X_train,y_train)
models['Linear Regression (all features)'] = lr1
```

```python
selected_feature = ["Age", "KM", "HP", "Weight", "Fuel_Petrol", "Fuel_Diesel"]
lr2 = LinearRegression()
lr2.fit(X_train[selected_feature], y_train)
models['Linear Regression (selected feature)'] = lr2
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly = PolynomialFeatures(degree=2)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
lr3 = LinearRegression()
lr3.fit(X_train_poly, y_train)
models['Linear Regression (Polynomial features)'] = lr3
```

```python
for name, model in models.items():
    if 'Polynomial' in name:
        y_pred = model.predict(X_test_poly)
    else:
        y_pred = model.predict(X_test[selected_feature] if 'selected' in name else X_test)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)
    print(f'{name} - MSE: {mse}, R2: {r2}')
```

```python
lasso = Lasso(alpha=0.1)
ridge = Ridge(alpha=1.0)
```

```python
lasso.fit(X_train, y_train)
y_pred_lasso = lasso.predict(X_test)
mse_lasso = mean_squared_error(y_test, y_pred_lasso)
r2_lasso = r2_score(y_test, y_pred_lasso)
print(f'Lasso Regression - MSE: {mse_lasso}, R2: {r2_lasso}')
```

```python
ridge.fit(X_train, y_train)
y_pred_ridge = ridge.predict(X_test)
mse_ridge = mean_squared_error(y_test, y_pred_ridge)
r2_ridge = r2_score(y_test, y_pred_ridge)
print(f'Ridge Regression - MSE: {mse_ridge}, R2: {r2_ridge}')
```

1.What is Normalization & Standardization and how is it helpful?

Ans

Normalization: Scales data to a range of [0, 1] using the formula: xnew=xmaxxminxxmin

Use: When features have different units or scales. Pros: Keeps data within a bounded range. Cons: Sensitive to outliers.

Standardization: Transforms data to have a mean of 0 and a standard deviation of 1 using the formula: xnew=x

Use: When data follows a Gaussian distribution. Pros: Less sensitive to outliers. Cons: Does not bound data within a specific range.

Both techniques help improve model performance, ensure features contribute equally, and aid in faster convergence of algorithms.

2.What techniques can be used to address multicollinearity in multiple linear regression?

Ans

To address multicollinearity in multiple linear regression, you can use several techniques:

Remove Highly Correlated Predictors: Identify and remove one of the highly correlated predictors to reduce redundancy.
Principal Component Analysis (PCA): Transform the predictors into a set of uncorrelated components.
Ridge Regression: Apply regularization to shrink coefficients and reduce multicollinearity.
Lasso Regression: Similar to Ridge, but can also set some coefficients to zero, effectively performing feature selection.
Variance Inflation Factor (VIF): Calculate VIF for each predictor and remove those with high VIF values (typically above 10).
Combine Predictors: Create a single predictor from highly correlated variables, such as averaging them.
