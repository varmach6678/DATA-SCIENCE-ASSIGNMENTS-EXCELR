```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, accuracy_score, confusion_matrix
```

```python
glass = df.copy()
```

```python
glass.hist(bins=15, figsize=(15, 10))
plt.suptitle("Histogram of Features", fontsize=15)
plt.show()

plt.figure(figsize=(12, 8))
sns.boxplot(data=glass.drop(columns='Type'))
plt.title('Box Plot of Features')
plt.show()

sns.pairplot(glass, hue='Type')
plt.show()

plt.figure(figsize=(10, 7))
sns.heatmap(glass.corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
print(glass.isnull().sum())
scaler = StandardScaler()
features = glass.drop(columns='Type')
scaled_features = scaler.fit_transform(features)
scaled_glass = pd.DataFrame(scaled_features, columns=features.columns)
scaled_glass['Type'] = glass['Type']
print(scaled_glass['Type'].value_counts())
```

```python
# Split the data into train and test sets
X = scaled_glass.drop(columns='Type')
y = scaled_glass['Type']

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

# Implement Random Forest Classifier
rfc = RandomForestClassifier(n_estimators=100, random_state=42)
rfc.fit(X_train, y_train)
y_pred = rfc.predict(X_test)

# Evaluate model performance
print("Accuracy:", accuracy_score(y_test, y_pred))
print("Classification Report:\n", classification_report(y_test, y_pred))

# Confusion matrix
conf_matrix = confusion_matrix(y_test, y_pred)
plt.figure(figsize=(7, 5))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.ensemble import BaggingClassifier

# Bagging Classifier with base model as Random Forest
bagging = BaggingClassifier(base_estimator=RandomForestClassifier(), n_estimators=100, random_state=42)
bagging.fit(X_train, y_train)

# Evaluate Bagging Model
y_pred_bagging = bagging.predict(X_test)
print("Bagging Accuracy:", accuracy_score(y_test, y_pred_bagging))
print("Bagging Classification Report:\n", classification_report(y_test, y_pred_bagging))
```

```python
from sklearn.ensemble import AdaBoostClassifier

# AdaBoost Classifier with base model as Random Forest
boosting = AdaBoostClassifier(base_estimator=RandomForestClassifier(), n_estimators=100, random_state=42)
boosting.fit(X_train, y_train)

# Evaluate Boosting Model
y_pred_boosting = boosting.predict(X_test)
print("Boosting Accuracy:", accuracy_score(y_test, y_pred_boosting))
print("Boosting Classification Report:\n", classification_report(y_test, y_pred_boosting))
```

1.Explain Bagging and Boosting methods. How is it different from each other.


Ans



Bagging is a method in ensemble learning that helps improve the accuracy and robustness of machine learning models. It does this by reducing variance through the combination of multiple models (usually the same model type) that are trained independently on different random subsets of the training data.
Boosting is another ensemble technique, but unlike Bagging, it focuses on sequentially building models, where each new model attempts to correct the errors made by previous models. Boosting improves accuracy by reducing bias and learning from mistakes.


2.Explain how to handle imbalance in the data.

Ans

Handling imbalanced datasets is crucial in machine learning because


it can lead to biased models that perform well on the majority class but poorly on the minority class. Here are several strategies to address class imbalance:

Resampling Techniques a. Oversampling the Minority Class
This involves increasing the number of instances in the minority class by duplicating existing samples or generating synthetic data. SMOTE (Synthetic Minority Over-sampling Technique) is a popular method that generates new, synthetic examples by interpolating between existing minority class samples. b. Undersampling the Majority Class

This technique reduces the number of instances in the majority class by randomly removing samples. While this can help balance the dataset, it may result in loss of important information. c. Combination of Over and Undersampling

Sometimes a combination of both techniques is effective, where some samples from the majority class are removed while others from the minority class are generated.

Using Different Evaluation Metrics Instead of accuracy, consider using metrics that give a better understanding of model performance on imbalanced data, such as: Precision: The ratio of true positives to the total predicted positives. Recall (Sensitivity): The ratio of true positives to the total actual positives. F1-Score: The harmonic mean of precision and recall, balancing both metrics. ROC-AUC: The area under the receiver operating characteristic curve provides a measure of performance across different thresholds.
Cost-Sensitive Learning Assign different misclassification costs to different classes, making the model more sensitive to the minority class. This can be done by modifying the loss function of the learning algorithm to penalize misclassifications of the minority class more heavily.
Ensemble Methods Use ensemble techniques that can help mitigate imbalance: Balanced Random Forest: A variant of Random Forest that balances the class distribution in each bootstrap sample. EasyEnsemble and BalanceCascade: These methods combine undersampling and boosting.
Anomaly Detection Approaches Treat the minority class as anomalies or outliers and apply anomaly detection techniques that can better identify these rare instances.
Synthetic Data Generation Use generative models (like GANs) to create synthetic samples for the minority class. This can be especially useful in complex scenarios.
Data Augmentation For certain types of data, such as images or text, data augmentation techniques can be applied to create variations of the existing minority class samples, thereby increasing their representation.
