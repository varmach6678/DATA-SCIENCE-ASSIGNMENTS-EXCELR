```python
# Import necessary libraries
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

# Ensure Matplotlib inline for Colab
%matplotlib inline

# Example DataFrame (Replace with your actual data)
data = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri', 'Mon', 'Tue'],
    'SKU': ['A1', 'A2', 'A1', 'A3', 'A2', 'A1', 'A3'],
    'City': ['NY', 'LA', 'NY', 'NY', 'LA', 'LA', 'NY'],
    'BU': ['Retail', 'Wholesale', 'Retail', 'Retail', 'Wholesale', 'Retail', 'Wholesale'],
    'Brand': ['X', 'Y', 'X', 'Z', 'Y', 'X', 'Z']
})

# Define categorical columns
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand']

# Set Seaborn style for consistent visuals
sns.set_style("whitegrid")

# Loop through each categorical column and plot bar charts
for column in categorical_columns:
    if column in data.columns:  # Check if column exists
        plt.figure(figsize=(8, 5))
        sns.countplot(
            y=data[column],
            order=data[column].value_counts().index,
            palette="viridis"
        )
        plt.title(f"Bar Chart of {column}")
        plt.ylabel(column)
        plt.xlabel("Frequency")
        plt.grid(axis='x')
        plt.show()
    else:
        print(f"Column '{column}' not found in the DataFrame.")
```

```python
import pandas as pd

# Example DataFrame (Replace with your actual data)
data = pd.DataFrame({
    'Volume': [100, 200, 150, 300],
    'Avg Price': [10.5, 12.0, 9.8, 15.0],
    'Total Sales Value': [1050, 2400, 1470, 4500],
    'Discount Rate (%)': [5, 10, 7, 12],
    'Discount Amount': [50, 240, 102.9, 540],
    'Net Sales Value': [1000, 2160, 1367.1, 3960]
})

# Define numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Calculate descriptive statistics
for column in numerical_columns:
    if column in data.columns:  # Check if column exists
        if pd.api.types.is_numeric_dtype(data[column]):  # Check if column is numeric
            mean = data[column].mean()
            median = data[column].median()
            mode = data[column].mode().iloc[0]  # Safe access to mode
            std_dev = data[column].std()

            print(f"{column}:\n")
            print(f"  Mean: {mean}")
            print(f"  Median: {median}")
            print(f"  Mode: {mode}")
            print(f"  Standard Deviation: {std_dev}\n")
        else:
            print(f"Column '{column}' is not numeric. Skipping.\n")
    else:
        print(f"Column '{column}' not found in the DataFrame. Skipping.\n")
```

```python
import pandas as pd

# Example DataFrame (Replace with your actual data)
data = pd.DataFrame({
    'Volume': [100, 200, 150, 300],
    'Avg Price': [10.5, 12.0, 9.8, 15.0],
    'Total Sales Value': [1050, 2400, 1470, 4500],
    'Discount Rate (%)': [5, 10, 7, 12],
    'Discount Amount': [50, 240, 102.9, 540],
    'Net Sales Value': [1000, 2160, 1367.1, 3960]
})

# Define numerical columns
numerical_columns = ['Volume', 'Avg Price', 'Total Sales Value', 'Discount Rate (%)',
                     'Discount Amount', 'Net Sales Value']

# Standardization using z-score normalization
standardized_data = data[numerical_columns].apply(lambda x: (x - x.mean()) / x.std())

# Comparison before and after standardization for one column
comparison_df = pd.DataFrame({
    'Original': data['Volume'],
    'Standardized': standardized_data['Volume']
})

# Print results
print("Comparison of Original and Standardized 'Volume':")
print(comparison_df)
```

```python
import pandas as pd

# Example DataFrame (Replace with your actual data)
data = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu'],
    'SKU': ['A1', 'A2', 'A1', 'A3'],
    'City': ['NY', 'LA', 'NY', 'LA'],
    'BU': ['Retail', 'Wholesale', 'Retail', 'Wholesale'],
    'Brand': ['X', 'Y', 'X', 'Z'],
    'Volume': [100, 200, 150, 300]
})

# Define categorical columns
categorical_columns = ['Day', 'SKU', 'City', 'BU', 'Brand']

# Check if all columns in categorical_columns exist in the DataFrame
for column in categorical_columns:
    if column not in data.columns:
        print(f"Warning: Column '{column}' not found in the DataFrame.")

# Convert categorical columns to dummy variables (one-hot encoding)
encoded_data = pd.get_dummies(data[categorical_columns], drop_first=True)

# Combine encoded data with the original dataset (excluding original categorical columns)
encoded_data_combined = pd.concat([data.drop(columns=categorical_columns), encoded_data], axis=1)

# Print the transformed dataset
print("Portion of the Transformed Dataset with One-Hot Encoding:")
print(encoded_data_combined.head())
```
